```python
#Drop "schoolsup", "famsup", "paid", "activities" from the data frame
dataset.drop(columns=["schoolsup", "famsup", "paid", "activities"], inplace=True)
dataset
```

# **Data manipulation: Rename Column** - "old_name":"new_name"

```python
#Rename these variables:
  #famsize - FamilySize
  #MJob- MotherJob
  #FJob- FatherJob
  #Dalc- WorkdayAlcoholConsumption
  #Walc- WeekendAlcoholConsumption
  #romantic - RomanticRelationship
  #famrel - Family Relationship
rename_data= dataset.rename(
        columns = {'famsize':'FamilySize'
                   ,'Mjob':'MotherJob'
                   ,'Fjob':'FatherJob'
                   ,'Dalc':'WorkdayAlcoholConsumption'
                   ,'Walc':'WeekendAlcoholConsumption'
                   ,'romantic':'RomanticRelationship'
                   ,'famrel':'Family Relationship'
                   }
        )
rename_data
```

# **Data Manipulation: Sorting data values**

```python
sort_data=rename_data.sort_values('age')
sort_data
```

# **Data Manipulation: Selecting data with specific data types**

```python
select_data=sort_data.select_dtypes('int')
select_data
```
